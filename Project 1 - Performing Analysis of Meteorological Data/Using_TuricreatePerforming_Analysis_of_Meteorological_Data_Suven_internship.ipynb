## Remove useless columns

```
weather_data=weather_data.remove_columns(['Daily Summary','Wind Bearing (degrees)','Summary','Precip Type','Temperature (C)',
              'Wind Speed (km/h)','Visibility (km)','Pressure (millibars)'],inplace=True)
```

```
weather_data
```

```
weather_data['Apparent Temperature']=weather_data['Apparent Temperature (C)']
weather_data.remove_column('Apparent Temperature (C)',inplace=True)
```

```
def find_month(mon):
  b=[]
  c=[]
  d=[]
  for i in weather_data:
    a=int(i['Formatted Date'][5:7])
    if a==mon:
      b.append(i['Humidity'])
      c.append(i['Apparent Temperature'])
      d.append(i['Formatted Date'])
  return [b,c,d]
```

```
o=find_month(3)
p,q,r=o[0],o[1],o[2]
print(p)
print(q)
h={'Formated Date':r,'Humidity':p,'Apparent Temperature':q}
print(h)
```

```
k=tc.SFrame(h)
#tc.plot(,tc.SFrame(q),xlabel='Humidity',ylabel='Apparant Temperature',title='Humidity vs Apparent Temperature of March')
print(k)
```

```
tc.plot(k['Humidity'][:5000],k['Apparent Temperature'][:5000],title='Humidity vs Apparent Temperature of March',xlabel='Humidity', ylabel='Apparant Temperature')
```

```
plt.figure(figsize=(15,5))
plt.plot(, marker='o', linestyle='-',label='Apparent Temperature (C)',color = 'green');
plt.plot(, marker='o', linestyle='-',label='Humidity',color = 'orange');
plt.legend(loc = 'center right',fontsize = 15);
plt.xlabel('Month of June', fontsize = 15);
plt.title('Humidity v/s Apparent Temperature',fontsize = 15)
```
